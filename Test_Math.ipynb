### Importation des bibliothèques

```python
import matplotlib.pyplot as plt
import numpy
```

### Définition de la fonction

```python
f= lambda x_values : numpy.exp(-x_values/10)*numpy.sin(x_values)
```

### Définition de la fonction dérivée

```python
# création de la fonction derivée avec la difference quotient

def df(x):
    h = 1e-5
    return (f(x+h)-f(x))/h
```

### Création de la fonction f et la fonction dérivée df

```python
a=0
b=10
number_points = 10000

x_values = numpy.linspace(a,b, number_points)
y_values = f(x_values)
y_values_prime = df(x_values)
```

# Question 1 et 2 :

### Définition de la fonction draw

```python
def draw_my_functions(f,start,end,number_of_points):
    
    
    x_values = numpy.linspace(start,end,number_of_points)
    y_values = f(x_values)
    y_values_prime = numpy.gradient(y_values,x_values)
    
    plt.figure(figsize=(15,6))
    plt.title("f et df")
    plt.plot(x_values,y_values)
    plt.plot(x_values,y_values_prime)
    plt.show()
```

### Graphe de f et df

```python
draw_my_functions(f,a,b,number_points)
```

### Calcul de la valeur moyenne et l'écart type

```python
'''
la méthode numpy.logical_and retourne un array des booléens. les paramétres de la fonction sont les conditions pour créer
un nouveau array qui contient les valeurs de x entre 3 et 7 
'''

x_values_new= x_values[numpy.logical_and(x_values>=3 ,x_values<=7)]


avg_moy = numpy.mean(f(x_values_new))
std_value = numpy.std(f(x_values_new))
print("la Valeur Moyenne est : ",avg_moy)
print("l'écart type est : ",std_value)
```

### Calcul de f'(x)=0 avec la dichotomie

```python
def dichotomie(f, a, b):
    eps= 1e-6
    while (b - a) > eps:
        m = (a + b) / 2
        if f(a) * f(m) < 0:
            b = m
        else:
            a = m
    return (a + b) / 2
            
```

```python
# d'aprés le grpahe de la fonction dérivée, on constitue que f'(x)=0 accepte 3 solutions

x1 = dichotomie(df, 0, 3)
x2 = dichotomie(df, 3, 6)
x3 = dichotomie(df, 6, 10)
print("Les solutions de f'(x) = 0 sont :", "x1 =",x1,"x2 =", x2, "x3 =", x3)
```
